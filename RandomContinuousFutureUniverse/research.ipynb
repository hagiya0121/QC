```python
from AlgorithmImports import *
import inspect

sector_symbols = {}

for sector_name, sector_class in inspect.getmembers(Futures):
    if inspect.isclass(sector_class) and sector_name[0].isupper():
        seen = set()
        symbols = []
        for name, value in inspect.getmembers(sector_class):
            if isinstance(value, str) and not name.startswith("_"):
                if value not in seen:
                    symbols.append((name, value))
                    seen.add(value)
        if symbols:
            sector_symbols[sector_name] = symbols

for sector, tickers in sector_symbols.items():
    print(f"\n=== {sector} ===")
    for name, ticker in tickers:
        print(f"{name}: {ticker}")

all_symbols = []
seen = set()
for sector, tickers in sector_symbols.items():
    for name, ticker in tickers:
        all_symbols.append((name, ticker))
        seen.add(ticker)

print(f"全銘柄数: {len(all_symbols)}")
```
